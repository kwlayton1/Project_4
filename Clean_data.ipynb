```python
# import the dependencies

import pandas as pd
```

```python
dataset=pd.read_csv("Resources/column_2C_weka.csv")
dataset.head()
```

```python
dataset.info()
```

```python
dataset.columns
```

```python
dataset=dataset.rename(columns={"pelvic_incidence":"Pelvic Incidence",
                                "pelvic_tilt numeric":"Pelvic Tilt",
                                "lumbar_lordosis_angle":"Lumbar Lordosis Angle",
                                "sacral_slope":"Sacral Slope",
                                "pelvic_radius":"Pelvic Radius",
                                "degree_spondylolisthesis":"Grade of spondylolisthesis",
                                "class":"Results"})
dataset.head()
```

```python
dataset["Pelvic Incidence"]=round(dataset["Pelvic Incidence"],2)
dataset.head()
```

```python
dataset["Pelvic Tilt"]=round(dataset["Pelvic Tilt"],2)
dataset["Lumbar Lordosis Angle"]=round(dataset["Lumbar Lordosis Angle"],2)
dataset["Sacral Slope"]=round(dataset["Sacral Slope"],2)
dataset["Pelvic Radius"]=round(dataset["Pelvic Radius"],2)
dataset["Grade of spondylolisthesis"]=round(dataset["Grade of spondylolisthesis"],2)
dataset.head()
```

```python
dataset.to_csv("Resources/dataset_clean.csv",index=False)
```
